```
import pyspark.sql.functions as F

assert output.count() == 5
assert output.filter(F.col("OrderId") == "197ef1bf-2d4d-4813-83ca-000037f8ffb6").count() == 2, "order mapping to two details"
assert output.filter(F.col("OrderId") == "1a8ccf3f-44b2-4f91-8a93-0000447a035c").count() == 1, "order mapping to one details"
assert output.filter(F.col("OrderId") == "3c326201-a86e-4083-a124-0000b3850edf").count() == 1, "order mapping to zero details"
assert output.filter(F.col("OrderId").isNull()).count() == 0, "detail mapping to zero order should fill OrderId"
assert output.filter(F.col("OrderDate").isNull()).count() == 1, "detail mapping to zero order should not fill OrderDate"

assert all(
   [row["n"] == 2 for row in output
    .filter(F.col("OrderId") == "197ef1bf-2d4d-4813-83ca-000037f8ffb6")
    .selectExpr("LinesInOrder as n")
    .collect()
   ]), "LinesInOrder for order mapping to two details"

```
